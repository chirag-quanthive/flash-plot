````python
from flash_plot import FlashPlot
import numpy as np

print(f"FlashPlot ready")
````

---
## 1. Line Chart — Cumulative Returns

````python
np.random.seed(42)
strategy = np.cumsum(np.random.randn(40) * 5).tolist()
benchmark = np.cumsum(np.random.randn(40) * 2).tolist()

fig = FlashPlot(width=595, height=300)
fig.set_title("Cumulative Returns")
fig.set_subtitle("Strategy vs Benchmark — 40 trading days")
fig.plot(strategy, color="#d4d4d4", label="Strategy")
fig.plot(benchmark, color="#707070", label="Benchmark")
fig.set_xticks(["W1", "W2", "W3", "W4", "W5", "W6", "W7", "W8"])
fig.grid(True)
fig.legend()
fig.show()
````

## 2. Grouped Bar Chart — Sector Allocation

````python
fig = FlashPlot(height=300)
labels = ["Tech", "Health", "Finance", "Energy", "Consumer", "Industrial"]
current = [28, 18, 15, 12, 14, 13]
target = [22, 20, 18, 10, 16, 14]

fig.set_title("Sector Allocation")
fig.set_subtitle("Portfolio weight (%)")
fig.bar(labels, current, label="Current")
fig.bar(labels, target, label="Target")
fig.grid(True)
fig.legend()
fig.show()
````

## 3. Scatter Plot — Risk vs Return

````python
np.random.seed(123)
vol = np.random.uniform(3, 20, 25)
ret = vol * 0.6 + np.random.randn(25) * 3

fig = FlashPlot(height=300)
fig.set_title("Risk vs Return")
fig.set_subtitle("Annualized metrics — 25 funds")
fig.scatter(vol, ret, color="#4ECDC4", label="Funds", size=6)
fig.axhline(0, color="#707070", line_style="dashed")
fig.grid(True)
fig.legend()
fig.show()
````

## 4. Histogram — Return Distribution

````python
np.random.seed(7)
returns = np.random.randn(500) * 2 + 0.5

fig = FlashPlot(height=280)
fig.set_title("Return Distribution")
fig.set_subtitle("Daily returns — 500 observations")
fig.hist(returns, bins=25, label="Returns")
fig.axvline(0, color="#FF6B6B", line_style="dashed")
fig.grid(True)
fig.legend()
fig.show()
````

## 5. Area Chart — AUM Growth

````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
aum = [100, 120, 115, 140, 160, 155, 180, 200, 195, 220, 250, 270]
x = list(range(len(aum)))

fig = FlashPlot(height=280)
fig.set_title("AUM Growth")
fig.set_subtitle("Assets under management ($M)")
fig.plot(aum, color="#C084FC", label="AUM")
fig.fill_between(x, aum, 0, color="#C084FC", alpha=0.12)
fig.set_xticks(months)
fig.grid(True)
fig.legend()
fig.show()
````

## 6. Multi-Line with Reference Lines

````python
np.random.seed(99)
alpha = np.cumsum(np.random.randn(60) * 1.5 + 0.3).tolist()
beta = np.cumsum(np.random.randn(60) * 1.0).tolist()
gamma = np.cumsum(np.random.randn(60) * 0.8 - 0.1).tolist()

fig = FlashPlot(width=650, height=320)
fig.set_title("Strategy Comparison")
fig.set_subtitle("60-day cumulative P&L")
fig.plot(alpha, color="#4ECDC4", label="Alpha", line_width=1.8)
fig.plot(beta, color="#C084FC", label="Beta")
fig.plot(gamma, color="#FFD93D", label="Gamma", line_style="dashed")
fig.axhline(0, color="#555", line_style="dotted")
fig.grid(True)
fig.legend()
fig.show()
````

## 7. Bar Chart — Monthly P&L

````python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
pnl = [12, -5, 18, 8, -3, 25, 15, -8, 20, 30, -2, 14]

fig = FlashPlot(height=280)
fig.set_title("Monthly P&L")
fig.set_subtitle("Strategy returns ($K)")
fig.bar(months, pnl, label="P&L", color="#4aaaba")
fig.axhline(0, color="#555", line_style="solid")
fig.grid(True)
fig.legend()
fig.show()
````
